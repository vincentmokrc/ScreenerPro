```python
import xml.etree.ElementTree as ET
import re
root = ET.parse('C:/Users/wsmok/Documents/Scrapy/LHI/LHIQ12019.pdf.xml').getroot()
wroot = ET.Element('Pdf')
regexes = ["Revenue",
            "Other income",
            "Operating expenses",
            "Profit from operations",
            "Finance costs",
            "Share of profits of associates",
            "Profit before tax",
            "Tax expense",
            "Net profit for the financial period",
            "Currency translation differences"]

combined = "(" + ")|(".join(regexes) + ")"
```

```python
for page in root.findall('page'):
    pagenumber = page.get('number')
    sub = ET.SubElement(wroot, 'Page', Number = pagenumber)
    for item in page.findall('text'):
        if(item.text is not None and re.search(r"[\d]{1,2}/[\d]{1,2}/[\d]{4}", item.text.strip())):
            position = int(item.get('left')) + int(item.get('width'))
            ET.SubElement(sub, 'Column', Position = str(position)).text = item.text
        if(item.text is not None and re.search(combined, item.text.strip())):
            ET.SubElement(sub, 'Row', Position = item.get('top')).text = item.text
```
