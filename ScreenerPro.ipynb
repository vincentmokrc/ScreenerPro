```python
import urllib.request
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd
```

```python
listing_directory = "https://www.bursamalaysia.com/trade/trading_resources/listing_directory/"
main_market = "main_market"
ace_market = "ace_market"
leap_market = "leap_market"
lfx_market = "lfx_market"
pn17_and_gn13_companies = "pn17_and_gn13_companies"
change_of_name = "change_of_name"
slash = "/"
```

```python
url_is = listing_directory + ace_market
```

```python
driver = webdriver.Firefox()
```

```python
driver.get(url_is)
```

```python
tutorial_soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
table = tutorial_soup.find('table', id="DataTables_Table_0")
```

```python
table_rows = table.find_all('tr')
```

```python
res = []
for tr in table_rows:
    td = tr.find_all('td')
    td2 = tr.find_all('td')[1:2]
    for d in td2: 
        a = d.find_all('a', href=True)
        for b in a:
            code=b['href'].split('stock_code=')[1]
            print(code)
            
    row = [d.text.strip() for d in td]
    row.append(code)
    if row:
        res.append(row)
```

```python
df = pd.DataFrame(res, columns=["No","Company Name", "Company Website","Code"])
```
